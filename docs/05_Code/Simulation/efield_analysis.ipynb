## COMSOL Lead Field → 전기장 크기 계산

### 1. 데이터 로드

KDY `load_leadfield_data` 기반 — 2D(Ex, Ey)용으로 수정.

반환 형태:
- `coords`: `(N, 2)` — x, y 좌표 [mm]
- `E_tensor`: `(num_electrodes, N, 2)` — 각 전극의 [Ex, Ey] [V/m, 실수부]

```python
def load_leadfield_2d(filepath, x_range=None, y_range=None):
    """
    COMSOL 2D lead field CSV 로드.
    - x_range/y_range=None 이면 데이터에서 도메인 자동 판단
    - NaN 행 제거
    - 준정적 전기장이므로 실수부만 사용 (허수부 ~1.5% 무시)
    반환: coords (N, 2), E_tensor (num_electrodes, N, 2)
    """
    print(f"Loading: {Path(filepath).name}")
    df = pd.read_csv(filepath, comment='%', header=None)

    # NaN 행 제거
    n_before = len(df)
    df = df.dropna().reset_index(drop=True)
    if len(df) < n_before:
        print(f"  [!] NaN 행 제거: {n_before - len(df)}개")

    cx = df.iloc[:, 0].values.astype(float)
    cy = df.iloc[:, 1].values.astype(float)

    # 도메인 자동 판단 (None이면 데이터 범위 그대로 사용)
    if x_range is None:
        x_range = (float(cx.min()), float(cx.max()))
    if y_range is None:
        y_range = (float(cy.min()), float(cy.max()))
    print(f"  도메인 — x: [{x_range[0]:.2f}, {x_range[1]:.2f}] mm  "
          f"y: [{y_range[0]:.2f}, {y_range[1]:.2f}] mm")

    # 좌표 범위 필터링
    mask = (
        (cx >= x_range[0]) & (cx <= x_range[1]) &
        (cy >= y_range[0]) & (cy <= y_range[1])
    )
    if (~mask).sum():
        print(f"  [!] 범위 밖 노드 제거: {(~mask).sum()}개")
    df = df[mask].reset_index(drop=True)

    coords     = df.iloc[:, :2].values.astype(float)
    field_cols = df.iloc[:, 2:]
    num_electrodes = field_cols.shape[1] // 2
    num_points     = len(df)

    # 복소수 파싱: 'i' → 'j', 실수부 추출
    parsed = np.zeros((num_points, field_cols.shape[1]), dtype=np.float64)
    for i in range(field_cols.shape[1]):
        col_str = (field_cols.iloc[:, i]
                   .fillna('0').astype(str)
                   .str.replace(' ', '', regex=False)
                   .str.replace('i', 'j', regex=False))
        parsed[:, i] = np.real(np.array([complex(v) for v in col_str]))

    E_tensor = np.zeros((num_electrodes, num_points, 2))
    for e in range(num_electrodes):
        E_tensor[e, :, 0] = parsed[:, e * 2]
        E_tensor[e, :, 1] = parsed[:, e * 2 + 1]

    print(f"  → 전극 {num_electrodes}개 / 유효 포인트 {num_points}개 / E_tensor: {E_tensor.shape}")
    return coords, E_tensor


coords, E_f1 = load_leadfield_2d(LEADFIELD_F1)
_,      E_f2 = load_leadfield_2d(LEADFIELD_F2)

x, y = coords[:, 0], coords[:, 1]
print(f"\n좌표 범위 — x: [{x.min():.2f}, {x.max():.2f}] mm  "
      f"y: [{y.min():.2f}, {y.max():.2f}] mm")
```

### 2. Anti-phase 채널 전기장 합성 & 크기 계산

각 파일의 두 전극(idx=1, idx=2)이 서로 **anti-phase** 쌍이므로 차분으로 채널 전기장 합성. (KDY 코드 루프 내 `Ef = E[a] - E[b]` 와 동일한 원리.)

$$\mathbf{E}_{ch} = \mathbf{E}_{idx=1} - \mathbf{E}_{idx=2}$$

$$|\mathbf{E}| = \sqrt{E_x^2 + E_y^2}$$

> **Note:** COMSOL Terminal이 이미 `1[mA]`로 계산 → CSV 값이 V/m @ 1mA.  별도 전류 스케일링 불필요.

```python
# ── Anti-phase 채널 전기장 합성 ───────────────────────────────────────────────
# COMSOL Terminal = 1[mA]로 계산 → CSV already in V/m @ 1mA → 별도 스케일 불필요
Ef1 = E_f1[0] - E_f1[1]   # (N, 2): [Ex_f1, Ey_f1] [V/m]
Ef2 = E_f2[0] - E_f2[1]   # (N, 2): [Ex_f2, Ey_f2] [V/m]

print(f"Ef1 shape: {Ef1.shape}  (채널 1: 전극 1,2 / {CURRENT_F1_MA} mA)")
print(f"Ef2 shape: {Ef2.shape}  (채널 2: 전극 3,4 / {CURRENT_F2_MA} mA)")

# ── 전기장 크기 계산 ──────────────────────────────────────────────────────────
E_mag_f1 = np.linalg.norm(Ef1, axis=1)   # (N,) [V/m]
E_mag_f2 = np.linalg.norm(Ef2, axis=1)   # (N,) [V/m]

print("\n── |E_f1| 통계 (V/m) ──")
print(f"  max={E_mag_f1.max():.6f}  mean={E_mag_f1.mean():.6f}  median={np.median(E_mag_f1):.6f}")
print("── |E_f2| 통계 (V/m) ──")
print(f"  max={E_mag_f2.max():.6f}  mean={E_mag_f2.mean():.6f}  median={np.median(E_mag_f2):.6f}")
```

### 3. AM Modulation 계산 (Grossman 2017)

두 채널 전기장이 조직 내에서 간섭할 때 생기는 저주파 envelope.

$$AM_{max} = \max_{\hat{n}} \left[ 2 \cdot \min\left(|\mathbf{E}_{f1} \cdot \hat{n}|,\ |\mathbf{E}_{f2} \cdot \hat{n}|\right) \right]$$

`compute_AM_Grossman` — KDY `KDY_optimization_electrode_origin.ipynb` 에서 가져옴 (2D 그대로 적용 가능).

```python
def compute_AM_Grossman(Ef1, Ef2):
    """
    Grossman 2017 AM modulation 계산.
    Ef1, Ef2: shape (N, 2) or (N, 3) — 각 채널의 전기장 벡터
    반환: AM_vector (N,) — 각 포인트의 AM envelope 크기 [V/m]
    """
    eps = 1e-15

    n1 = np.linalg.norm(Ef1, axis=1)
    n2 = np.linalg.norm(Ef2, axis=1)
    n1 = np.where(n1 == 0, eps, n1)
    n2 = np.where(n2 == 0, eps, n2)

    # 크기 기준으로 E_max / E_min 분류
    cond   = n1 > n2
    E_max  = np.where(cond[:, None], Ef1, Ef2)
    E_min  = np.where(cond[:, None], Ef2, Ef1)
    E1_amp = np.where(cond, n1, n2)
    E2_amp = np.where(cond, n2, n1)

    # 두 벡터 사이 각도
    dot_prod = np.sum(E_max * E_min, axis=1)
    cosB     = np.clip(dot_prod / (E1_amp * E2_amp), -1.0, 1.0)
    ang      = np.arccos(cosB)

    # 예각/둔각 분기
    acute_cond  = ang < (np.pi / 2)
    E_min_adj   = np.where(acute_cond[:, None], E_min, -E_min)
    cosA        = np.where(acute_cond, cosB, -cosB)

    E_m      = E_max - E_min_adj
    norm_E_m = np.linalg.norm(E_m, axis=1)
    norm_E_m = np.where(norm_E_m == 0, eps, norm_E_m)

    dot_max_min_adj = np.sum(E_max * E_min_adj, axis=1)
    cosBB = np.clip((-(E2_amp**2) + dot_max_min_adj) / (E2_amp * norm_E_m), -1.0, 1.0)
    sinBB = np.sin(np.arccos(cosBB))

    am_cond   = E2_amp < (E1_amp * cosA)
    AM_vector = np.where(am_cond, 2 * E2_amp, 2 * E2_amp * sinBB)

    return AM_vector


AM_max = compute_AM_Grossman(Ef1, Ef2)   # (N,) [V/m]

print("── AM_max 통계 (V/m) ──")
print(f"  max={AM_max.max():.8f}  mean={AM_max.mean():.8f}  median={np.median(AM_max):.8f}")


df_result = pd.DataFrame({
    "x (mm)"       : x,
    "y (mm)"       : y,
    "Ex_f1 (V/m)"  : Ef1[:, 0],
    "Ey_f1 (V/m)"  : Ef1[:, 1],
    "|E_f1| (V/m)" : E_mag_f1,
    "Ex_f2 (V/m)"  : Ef2[:, 0],
    "Ey_f2 (V/m)"  : Ef2[:, 1],
    "|E_f2| (V/m)" : E_mag_f2,
    "AM_max (V/m)" : AM_max,
})

#df_result.sort_values("AM_max (V/m)", ascending=False).head(10).reset_index(drop=True)
df_result.describe()
```

### 4. 시각화 — 공간 분포 + X/Y축 단면 프로파일 + max-projection 프로파일 (동일 격자 기반)

- `griddata`로 **공통 2D 격자(matrix) `grid_E_f1`, `grid_E_f2`, `grid_AM`을 딱 한 번만 생성**하고,
  1행(2D map)·2행(고정선 프로파일)·3행(max-projection 프로파일)이 모두 이 행렬에서만 값을 가져온다.
- **1행**: |E_f1|, |E_f2|, AM_max의 2D 공간 분포 (`pcolormesh`, `FIX_COLORBAR`로 공통/자동 스케일 선택)
- **2행**: `FIX_Y`/`FIX_X`로 고정한 한 줄(row/col)만 그대로 슬라이스한 프로파일
- **3행 (신규)**: 각 x에서 y 전체 중 최댓값 / 각 y에서 x 전체 중 최댓값을 그린 **max-projection 프로파일** — 2행의 고정선 프로파일이 전극 근처(예: y≠0)의 실제 핫스팟 값을 놓치는 문제를, 축을 따라 이동하며 '그 축에서 도달 가능한 최댓값'을 보여줌으로써 보완한다. 1행 맵에서 보이는 빨간 핫스팟 값이 여기서는 실제 수치로 다시 확인된다.
- **검증 1~5**: 원본 mesh 노드 개수, 좌우 대칭성, f1/f2 대칭성, Grossman 상한 검증, 핫스팟-절단선 거리 비교 (기존과 동일, 콘솔 출력)

```python
# ── 전극 좌표 / 팬텀 geometry (시각화용, 사용자 실측값 mm) — 이 셀에서만 정의 ──
ELECTRODE_XY_TABLE = {
    (50, 50): (25, 20),
    (30, 50): (15, 20),
    (10, 50): (5, 20),
    (50, 30): (20, 15),
    (50, 10): (20, 5),
    (30, 30): (15, 10),
    (10, 10): (5, 2.5),
}
ELECTRODE_RADIUS_MM = 2.0   # 지름 2 mm

_xy_match = re.search(r"X(\d+)_Y(\d+)$", TAG)
if _xy_match is None:
    raise ValueError(f"TAG '{TAG}'에서 X##_Y## 패턴을 찾을 수 없습니다.")
PHANTOM_X_MM, PHANTOM_Y_MM = int(_xy_match.group(1)), int(_xy_match.group(2))

if (PHANTOM_X_MM, PHANTOM_Y_MM) not in ELECTRODE_XY_TABLE:
    raise KeyError(
        f"팬텀 크기 ({PHANTOM_X_MM}, {PHANTOM_Y_MM})에 대한 전극 좌표가 "
        f"ELECTRODE_XY_TABLE에 없습니다. 표를 갱신하세요."
    )
_ex, _ey = ELECTRODE_XY_TABLE[(PHANTOM_X_MM, PHANTOM_Y_MM)]

ELECTRODES = {
    "CH1 top": (-_ex,  _ey),
    "CH1 bot": (-_ex, -_ey),
    "CH2 top": ( _ex,  _ey),
    "CH2 bot": ( _ex, -_ey),
}

print(f"팬텀 크기: {PHANTOM_X_MM} x {PHANTOM_Y_MM} mm  /  전극 좌표(1사분면): ({_ex}, {_ey}) mm")

PAD_MM = 3.0
FRAME_X_HALF = max(PHANTOM_X_MM / 2, _ex + ELECTRODE_RADIUS_MM) + PAD_MM
FRAME_Y_HALF = max(PHANTOM_Y_MM / 2, _ey + ELECTRODE_RADIUS_MM) + PAD_MM


def draw_geometry_overlay(ax):
    """팬텀 경계(검은 실선 사각형) + 전극(회색 원) 오버레이. x/y축 범위는 건드리지 않음."""
    rect = plt.Rectangle(
        (-PHANTOM_X_MM / 2, -PHANTOM_Y_MM / 2), PHANTOM_X_MM, PHANTOM_Y_MM,
        fill=False, edgecolor="black", linewidth=1.8, linestyle="-", zorder=5,
    )
    ax.add_patch(rect)
    for name, (px, py) in ELECTRODES.items():
        circ = plt.Circle((px, py), ELECTRODE_RADIUS_MM,
                           facecolor="gray", edgecolor="black", linewidth=1.0, zorder=6)
        ax.add_patch(circ)


def apply_sci_colorbar(cb, vmin, vmax, power=-2, n_ticks=6,
                        label_fontsize=16, tick_fontsize=15):
    """Colorbar tick을 ×10^power 단위 숫자로 표시하고 label에 배율 표기를 덧붙인다."""
    scale = 10 ** power
    ticks = np.linspace(vmin, vmax, n_ticks)
    cb.set_ticks(ticks)
    cb.set_ticklabels([f"{t / scale:.2f}" for t in ticks])
    cb.ax.tick_params(labelsize=tick_fontsize)
    base_label = cb.ax.get_ylabel().split("  (×10")[0]
    cb.set_label(f"{base_label}  (×10$^{{{power}}}$)", fontsize=label_fontsize)


# ═══════════════════════════════════════════════════════════════════════════
# 공통 2D 격자(matrix) 생성 — 맵과 프로파일이 반드시 "같은 배열"에서 나오도록
# griddata를 여기서 딱 한 번만 호출하고, 맵은 이 행렬을 그대로 그리고
# 프로파일은 이 행렬에서 행/열을 슬라이스만 한다 (별도 재계산 없음).
# ═══════════════════════════════════════════════════════════════════════════
GRID_RES = 300
xi = np.linspace(x.min(), x.max(), GRID_RES)
yi = np.linspace(y.min(), y.max(), GRID_RES)
Xi, Yi = np.meshgrid(xi, yi)   # Xi, Yi shape: (GRID_RES, GRID_RES)

def _build_grid(values):
    """coords 위의 값을 (Xi, Yi) 정규 격자로 보간. convex hull 밖(NaN)은 nearest로 채움."""
    g = griddata(coords, values, (Xi, Yi), method="linear")
    nan_mask = np.isnan(g)
    if nan_mask.any():
        g_near = griddata(coords, values, (Xi, Yi), method="nearest")
        g[nan_mask] = g_near[nan_mask]
    return g

grid_E_f1 = _build_grid(E_mag_f1)   # (GRID_RES, GRID_RES) — 행: y, 열: x
grid_E_f2 = _build_grid(E_mag_f2)
grid_AM   = _build_grid(AM_max)

panels = [
    (f"|E$_{{f1}}$|  (f1: {CURRENT_F1_MA} mA)", grid_E_f1),
    (f"|E$_{{f2}}$|  (f2: {CURRENT_F2_MA} mA)", grid_E_f2),
    ("AM$_{max}$  (Grossman 2017)",             grid_AM),
]

# Colorbar 고정 여부 스위치
# True  : 아래 VMIN/VMAX를 모든 패널·모든 조건(TAG)에 공통으로 적용 → 조건 간 객관적 비교용
# False : 패널마다 자기 데이터의 [0, max]로 자동 스케일 → 이 조건 하나의 분포를 최대 대비로 보고 싶을 때
FIX_COLORBAR = False
VMIN, VMAX = 0, 0.05   # FIX_COLORBAR=True일 때만 사용되는 공통 range

if not FIX_COLORBAR:
    print("[!] FIX_COLORBAR=False — 패널마다 자체 데이터 범위로 자동 스케일 (조건 간 비교에는 부적합)")

# ═══════════════════════════════════════════════════════════════════════════
# X축 프로파일(y=FIX_Y 고정) / Y축 프로파일(x=FIX_X 고정)
#  — 위에서 만든 grid_E_f1 / grid_E_f2 / grid_AM "행렬"에서 행/열을 그대로 슬라이스.
#    FIX_Y/FIX_X와 정확히 같은 격자선이 없으면 가장 가까운 격자 인덱스를 사용한다.
# ═══════════════════════════════════════════════════════════════════════════
FIX_Y = 0   # X축 프로파일에서 고정할 y [mm]
FIX_X = 0   # Y축 프로파일에서 고정할 x [mm]
NODE_TOL_MM = 1.0   # 원본 mesh 노드 오버레이용 허용 오차 [mm]

idx_y = int(np.argmin(np.abs(yi - FIX_Y)))   # X축 프로파일이 사용할 행(row) 인덱스
idx_x = int(np.argmin(np.abs(xi - FIX_X)))   # Y축 프로파일이 사용할 열(col) 인덱스
actual_fix_y = float(yi[idx_y])
actual_fix_x = float(xi[idx_x])
print(f"[격자 슬라이스] X축 프로파일: 요청 y={FIX_Y} → 실제 격자선 y={actual_fix_y:.3f} mm (idx={idx_y})")
print(f"[격자 슬라이스] Y축 프로파일: 요청 x={FIX_X} → 실제 격자선 x={actual_fix_x:.3f} mm (idx={idx_x})")

x_line = xi
prof_x_E_f1 = grid_E_f1[idx_y, :]
prof_x_E_f2 = grid_E_f2[idx_y, :]
prof_x_AM   = grid_AM[idx_y, :]

y_line = yi
prof_y_E_f1 = grid_E_f1[:, idx_x]
prof_y_E_f2 = grid_E_f2[:, idx_x]
prof_y_AM   = grid_AM[:, idx_x]

# ── max-projection 프로파일 ───────────────────────────────────────────────
# X축으로 보되, 각 x에서 y 전체를 훑어 나온 최댓값(그 x열의 세로선 중 가장 큰 값)을 그린다.
# → 전극 근처(y가 0이 아닌 곳)의 진짜 핫스팟 크기가 x에 따라 어떻게 변하는지 보여준다.
maxproj_x_E_f1 = np.nanmax(grid_E_f1, axis=0)   # (GRID_RES,) — 열(=x)마다 행(=y) 최댓값
maxproj_x_E_f2 = np.nanmax(grid_E_f2, axis=0)
maxproj_x_AM   = np.nanmax(grid_AM,   axis=0)

# Y축으로 보되, 각 y에서 x 전체를 훑어 나온 최댓값(그 y행의 가로선 중 가장 큰 값)을 그린다.
maxproj_y_E_f1 = np.nanmax(grid_E_f1, axis=1)   # (GRID_RES,) — 행(=y)마다 열(=x) 최댓값
maxproj_y_E_f2 = np.nanmax(grid_E_f2, axis=1)
maxproj_y_AM   = np.nanmax(grid_AM,   axis=1)

# 검증 1 — 원본 mesh 노드 오버레이 (고정선 근방 실측값과 격자 슬라이스 비교)
near_x_mask = np.abs(y - actual_fix_y) < NODE_TOL_MM
near_y_mask = np.abs(x - actual_fix_x) < NODE_TOL_MM
print(f"[검증 1] X축 프로파일 — |y-{actual_fix_y:.2f}|<{NODE_TOL_MM}mm 인 실측 노드: {near_x_mask.sum()}개")
print(f"[검증 1] Y축 프로파일 — |x-{actual_fix_x:.2f}|<{NODE_TOL_MM}mm 인 실측 노드: {near_y_mask.sum()}개")

# 검증 2 — 좌우 대칭성 검사 (전극 배치가 x=0, y=0 축 대칭이므로 FIX≈0일 때 성립)
def _symmetry_check(t_line, prof, fix_val, axis_name):
    if abs(fix_val) > 1e-9:
        print(f"[검증 2] {axis_name}축 프로파일: FIX≠0 이라 좌우 대칭 전제가 성립하지 않음 — 대칭성 검사 생략")
        return
    prof_mirror = np.interp(-t_line, t_line, prof)
    max_abs_diff = np.max(np.abs(prof - prof_mirror))
    rel_diff_pct = 100 * max_abs_diff / (prof.max() + 1e-15)
    print(f"[검증 2] {axis_name}축 프로파일 좌우 대칭 오차: "
          f"max|prof(t)-prof(-t)|={max_abs_diff:.6f} V/m  ({rel_diff_pct:.2f}% of peak)")

_symmetry_check(x_line, prof_x_AM, actual_fix_y, "X")
_symmetry_check(y_line, prof_y_AM, actual_fix_x, "Y")

# 검증 3 — f1/f2 좌우 대칭 (전극쌍이 x=±ex에 대칭 배치일 때 x≈0이면 |E_f1|=|E_f2|)
if abs(actual_fix_x) < 1e-9:
    f1_f2_diff = np.max(np.abs(prof_y_E_f1 - prof_y_E_f2))
    f1_f2_rel  = 100 * f1_f2_diff / (max(prof_y_E_f1.max(), prof_y_E_f2.max()) + 1e-15)
    print(f"[검증 3] Y축 프로파일(x≈0) — |E_f1|와 |E_f2| 최대 차이: {f1_f2_diff:.6f} V/m "
          f"({f1_f2_rel:.2f}%)  ← 전극쌍이 x=0 기준 좌우 대칭이라 두 값이 이론적으로 같음")

# 검증 4 — AM_max가 개별 필드보다 커지는 지점 확인 (버그 아님: Grossman 공식상
#          |E1|≈|E2|인 지점에서 AM_max = 2*min(|E1|,|E2|)에 근접해 개별 필드보다 커질 수 있음)
peak_idx = int(np.argmax(prof_x_AM))
print(f"[검증 4] X축 프로파일 AM_max 최댓값 지점 (x={x_line[peak_idx]:.2f}mm): "
      f"|E_f1|={prof_x_E_f1[peak_idx]:.5f}, |E_f2|={prof_x_E_f2[peak_idx]:.5f}, "
      f"AM_max={prof_x_AM[peak_idx]:.5f}  "
      f"(AM_max ≤ 2*min(|E_f1|,|E_f2|)={2*min(prof_x_E_f1[peak_idx], prof_x_E_f2[peak_idx]):.5f} 이면 정상)")

# 검증 5 — 핫스팟 위치 vs 절단선 위치 비교
#          맵에서 보이는 "가장 빨간(전역 최댓값) 지점"이 프로파일 절단선(y=FIX_Y / x=FIX_X)에서
#          얼마나 떨어져 있는지 명시적으로 보여준다. 맵 색상이 진해 보여도 그 핫스팟이 절단선에서
#          멀리 있으면, 그 선을 따라 그린 1D 프로파일 값은 당연히 더 낮게 나온다 (계산 오류 아님).
def _report_hotspot(name, grid_mag, actual_fix_y, actual_fix_x):
    flat_idx = int(np.nanargmax(grid_mag))
    iy, ix = np.unravel_index(flat_idx, grid_mag.shape)
    hot_x, hot_y, hot_val = float(Xi[iy, ix]), float(Yi[iy, ix]), float(grid_mag[iy, ix])
    dist_from_xline = abs(hot_y - actual_fix_y)   # X축 프로파일선(y=actual_fix_y)까지 y방향 거리
    dist_from_yline = abs(hot_x - actual_fix_x)   # Y축 프로파일선(x=actual_fix_x)까지 x방향 거리
    print(f"[검증 5] {name} 전역 최댓값={hot_val:.5f} V/m @ (x={hot_x:.1f}, y={hot_y:.1f}) mm  "
          f"— X축 절단선(y={actual_fix_y:.1f})까지 거리={dist_from_xline:.1f}mm, "
          f"Y축 절단선(x={actual_fix_x:.1f})까지 거리={dist_from_yline:.1f}mm")

_report_hotspot("|E_f1|", grid_E_f1, actual_fix_y, actual_fix_x)
_report_hotspot("|E_f2|", grid_E_f2, actual_fix_y, actual_fix_x)
_report_hotspot("AM_max", grid_AM,   actual_fix_y, actual_fix_x)

# ═══════════════════════════════════════════════════════════════════════════
# 종합 시각화 — 2행: 위 3패널(2D map, pcolormesh) + 아래 2패널(X/Y 프로파일)
#   맵은 grid_E_f1/grid_E_f2/grid_AM을 pcolormesh로 그대로 그리고,
#   프로파일은 그 "같은 배열"의 행/열을 슬라이스한 값이므로 서로 완전히 일치한다.
# ═══════════════════════════════════════════════════════════════════════════
fig = plt.figure(figsize=(18, 15))
gs = gridspec.GridSpec(3, 6, figure=fig, hspace=0.5, wspace=0.9)

map_axes = [fig.add_subplot(gs[0, 0:2]), fig.add_subplot(gs[0, 2:4]), fig.add_subplot(gs[0, 4:6])]

for ax, (title, grid_mag) in zip(map_axes, panels):
    if FIX_COLORBAR:
        vmin, vmax = VMIN, VMAX
    else:
        vmin, vmax = 0, float(np.nanmax(grid_mag)*0.75)

    im = ax.pcolormesh(Xi, Yi, grid_mag, cmap="rainbow", vmin=vmin, vmax=vmax, shading="auto")
    cb = fig.colorbar(im, ax=ax, shrink=0.85)
    cb.set_label("|E| (V/m)", fontsize=14)
    apply_sci_colorbar(cb, vmin, vmax, label_fontsize=13, tick_fontsize=12)
    ax.set_title(title, fontsize=16)
    ax.set_xlabel("x (mm)", fontsize=13)
    ax.set_ylabel("y (mm)", fontsize=13)
    ax.set_aspect("equal")
    draw_geometry_overlay(ax)
    ax.axhline(actual_fix_y, color="black", lw=1.0, ls=":", alpha=0.7)   # X축 프로파일 절단선
    ax.axvline(actual_fix_x, color="black", lw=1.0, ls=":", alpha=0.7)   # Y축 프로파일 절단선
    ax.set_xlim(-FRAME_X_HALF, FRAME_X_HALF)
    ax.set_ylim(-FRAME_Y_HALF, FRAME_Y_HALF)
    ax.grid(True, color="white", alpha=0.4, linewidth=0.6, linestyle="--", zorder=4)

ax_px = fig.add_subplot(gs[1, 0:3])
ax_px.plot(x_line, prof_x_E_f1, color="tab:blue",   lw=2.0, ls="--", label="|E$_{f1}$|")
ax_px.plot(x_line, prof_x_E_f2, color="tab:orange", lw=2.0, ls="--", label="|E$_{f2}$|")
ax_px.plot(x_line, prof_x_AM,   color="tab:red",    lw=2.8,          label="AM$_{max}$ (Grossman)")
# if near_x_mask.sum() > 0:
#     ax_px.scatter(x[near_x_mask], AM_max[near_x_mask], color="black", s=14, zorder=5,
#                   label=f"실측 노드 (|y-{actual_fix_y:.1f}|<{NODE_TOL_MM}mm)")
ax_px.axvline(0, color="gray", lw=1.0, ls=":")
ax_px.set_xlabel(f"x (mm)  [y = {actual_fix_y:.2f} 고정, 격자 슬라이스]", fontsize=13)
ax_px.set_ylabel("Field magnitude (V/m)", fontsize=13)
ax_px.set_title(f"X축 프로파일 (y={actual_fix_y:.2f}) — 2D map과 동일 격자에서 추출", fontsize=13)
ax_px.legend(fontsize=10)
ax_px.grid(True, alpha=0.3)
ax_px.set_xlim(x_line.min(), x_line.max())

ax_py = fig.add_subplot(gs[1, 3:6])
ax_py.plot(y_line, prof_y_E_f1, color="tab:blue",   lw=5.0, alpha=0.45, solid_capstyle="round",
           label="|E$_{f1}$|", zorder=2)
ax_py.plot(y_line, prof_y_E_f2, color="tab:orange", lw=1.6, ls="--", label="|E$_{f2}$|", zorder=3)
ax_py.plot(y_line, prof_y_AM,   color="tab:red",    lw=2.8,          label="AM$_{max}$ (Grossman)", zorder=4)
# if near_y_mask.sum() > 0:
#     ax_py.scatter(y[near_y_mask], AM_max[near_y_mask], color="black", s=14, zorder=5,
#                   label=f"실측 노드 (|x-{actual_fix_x:.1f}|<{NODE_TOL_MM}mm)")
ax_py.axvline(0, color="gray", lw=1.0, ls=":")
ax_py.set_xlabel(f"y (mm)  [x = {actual_fix_x:.2f} 고정, 격자 슬라이스]", fontsize=13)
ax_py.set_ylabel("Field magnitude (V/m)", fontsize=13)
title_y = f"Y축 프로파일 (x={actual_fix_x:.2f}) — 2D map과 동일 격자에서 추출"
if abs(actual_fix_x) < 1e-9:
    title_y += "\n(|E$_{f1}$| ≈ |E$_{f2}$|: 좌우 대칭)"
ax_py.set_title(title_y, fontsize=12)
ax_py.legend(fontsize=10)
ax_py.grid(True, alpha=0.3)
ax_py.set_xlim(y_line.min(), y_line.max())

ax_mx = fig.add_subplot(gs[2, 0:3])
ax_mx.plot(x_line, maxproj_x_E_f1, color="tab:blue",   lw=2.0, ls="--", label="|E$_{f1}$| (y 최댓값)")
ax_mx.plot(x_line, maxproj_x_E_f2, color="tab:orange", lw=2.0, ls="--", label="|E$_{f2}$| (y 최댓값)")
ax_mx.plot(x_line, maxproj_x_AM,   color="tab:red",    lw=2.8,          label="AM$_{max}$ (y 최댓값)")
ax_mx.set_xlabel("x (mm)", fontsize=13)
ax_mx.set_ylabel("Field magnitude (V/m)", fontsize=13)
ax_mx.set_title("X축 — 각 x에서 y 전체 중 최댓값", fontsize=14)
ax_mx.legend(fontsize=10)
ax_mx.grid(True, alpha=0.3)
ax_mx.set_xlim(x_line.min(), x_line.max())

ax_my = fig.add_subplot(gs[2, 3:6])
ax_my.plot(y_line, maxproj_y_E_f1, color="tab:blue",   lw=2.0, ls="--", label="|E$_{f1}$| (x 최댓값)")
ax_my.plot(y_line, maxproj_y_E_f2, color="tab:orange", lw=2.0, ls="--", label="|E$_{f2}$| (x 최댓값)")
ax_my.plot(y_line, maxproj_y_AM,   color="tab:red",    lw=2.8,          label="AM$_{max}$ (x 최댓값)")
ax_my.set_xlabel("y (mm)", fontsize=13)
ax_my.set_ylabel("Field magnitude (V/m)", fontsize=13)
ax_my.set_title("Y축 — 각 y에서 x 전체 중 최댓값", fontsize=14)
ax_my.legend(fontsize=10)
ax_my.grid(True, alpha=0.3)
ax_my.set_xlim(y_line.min(), y_line.max())

fig.suptitle(
    f"Electric Field & AM Modulation — 공간 분포 + 단면 프로파일 (동일 격자)  [{TAG}]"
    + ("" if FIX_COLORBAR else "  [colorbar: 패널별 자동 스케일]"),
    fontsize=19, fontweight="bold"
)
plt.tight_layout()
_scale_tag = "fixed" if FIX_COLORBAR else "auto"
save_path = SIM_DIR / f"Result_Efield_map_and_profile_{TAG}_{_scale_tag}.png"
plt.savefig(save_path, dpi=150, bbox_inches="tight")
print(f"Saved: {save_path.resolve()}")
plt.show()
```

### 5. COMSOL 시뮬레이션 결과와 검증 (Point-level + Full-field)

COMSOL `AM_vector` surface export(`..._point_AM_vector.csv`, steady-state 전체 노드) 하나만 로드해서 Python lead field 기반 AM$_{max}$(Grossman 2017)와 **두 가지 방식**으로 비교한다. (과거에 쓰던 별도 sparse `*_point_ver2.csv` 검증 파일은 더 이상 필요 없음 — AM_vector 파일 하나로 두 분석 모두 수행)

- **방법 A (Point-level)**: AM_vector 파일의 좌표마다 `cKDTree`로 가장 가까운 Python leadfield 노드를 매칭 → 보간 없이 실제 노드 값끼리 직접 비교
- **방법 B (Full-field)**: 두 데이터셋을 공통 정규 격자에 `griddata` 보간 후 비교 → 공간 분포/패턴 비교

두 방법의 정확도(RMSE, MAE, MAPE, Accuracy=100-MAPE, R²)를 나란히 비교해 서로 일관되는지 교차 검증하고, parity plot과 오차 분포 히스토그램으로 세부적으로 점검한다.


```python
# ── COMSOL AM_vector surface export 로드 (한 번만 — point-level/full-field 공용) ──
df_comsol_am = pd.read_csv(AM_VECTOR_FILE, comment="%", header=None,
                            names=["x", "y", "AM_vector"])
comsol_xy = df_comsol_am[["x", "y"]].values
comsol_am = df_comsol_am["AM_vector"].values
print(f"COMSOL AM_vector 좌표 수: {len(comsol_am)}  범위: [{comsol_am.min():.4f}, {comsol_am.max():.4f}] V/m")

# ── lead field → Python AM_max (Grossman) ────────────────────────────────────
val_coords, val_Ef1_tensor = load_leadfield_2d(LEADFIELD_F1)
_,          val_Ef2_tensor = load_leadfield_2d(LEADFIELD_F2)
val_Ef1 = val_Ef1_tensor[0] - val_Ef1_tensor[1]
val_Ef2 = val_Ef2_tensor[0] - val_Ef2_tensor[1]
val_AM  = compute_AM_Grossman(val_Ef1, val_Ef2)


# ═══════════════════════════════════════════════════════════════════════════
# 방법 A — Point-level: AM_vector 좌표마다 nearest Python leadfield 노드 매칭
#          (mesh 대 mesh 직접 비교, 보간 없음 → 실제 값 오차)
# ═══════════════════════════════════════════════════════════════════════════
tree = cKDTree(val_coords)
nn_dist, nn_idx = tree.query(comsol_xy)

python_am_matched = val_AM[nn_idx]
diff_abs_pt = python_am_matched - comsol_am
diff_pct_pt = np.divide(
    np.abs(diff_abs_pt), np.abs(comsol_am),
    out=np.full_like(comsol_am, np.nan), where=np.abs(comsol_am) > 1e-9,
) * 100

df_val = pd.DataFrame({
    "x (mm)": comsol_xy[:, 0], "y (mm)": comsol_xy[:, 1],
    "COMSOL_AM (V/m)": comsol_am, "Python_AM (V/m)": python_am_matched,
    "diff (V/m)": diff_abs_pt, "diff (%)": diff_pct_pt, "nn_dist (mm)": nn_dist,
})

rmse_pt = np.sqrt(np.mean(diff_abs_pt ** 2))
mae_pt  = np.mean(np.abs(diff_abs_pt))
mape_pt = np.nanmean(diff_pct_pt)
acc_pt  = 100 - mape_pt
corr_pt = np.corrcoef(python_am_matched, comsol_am)[0, 1]
r2_pt   = corr_pt ** 2

print(f"\n[방법 A: Point-level, nearest-node]  n={len(df_val)}  "
      f"nn_dist max={nn_dist.max():.3f} mm  mean={nn_dist.mean():.3f} mm")
print(df_val["diff (%)"].describe())
print("\n오차 상위 10개 좌표:")
display(df_val.sort_values("diff (%)", ascending=False).head(10).round(6))


# ═══════════════════════════════════════════════════════════════════════════
# 방법 B — Full-field: 공통 정규 격자에 보간 후 비교 (공간 분포/패턴 비교용)
# ═══════════════════════════════════════════════════════════════════════════
GRID_RES = 200
xg_min = max(val_coords[:, 0].min(), comsol_xy[:, 0].min())
xg_max = min(val_coords[:, 0].max(), comsol_xy[:, 0].max())
yg_min = max(val_coords[:, 1].min(), comsol_xy[:, 1].min())
yg_max = min(val_coords[:, 1].max(), comsol_xy[:, 1].max())

xi = np.linspace(xg_min, xg_max, GRID_RES)
yi = np.linspace(yg_min, yg_max, GRID_RES)
Xi, Yi = np.meshgrid(xi, yi)

python_grid = griddata(val_coords, val_AM, (Xi, Yi), method="linear")
comsol_grid = griddata(comsol_xy, comsol_am, (Xi, Yi), method="linear")
diff_grid   = python_grid - comsol_grid

valid_mask = ~(np.isnan(diff_grid) | np.isnan(python_grid) | np.isnan(comsol_grid))
p_valid, c_valid, d_valid = python_grid[valid_mask], comsol_grid[valid_mask], diff_grid[valid_mask]

rmse_gd = np.sqrt(np.mean(d_valid ** 2))
mae_gd  = np.mean(np.abs(d_valid))
nz      = np.abs(c_valid) > 1e-9
mape_gd = np.mean(np.abs(d_valid[nz]) / np.abs(c_valid[nz])) * 100
acc_gd  = 100 - mape_gd
corr_gd = np.corrcoef(p_valid, c_valid)[0, 1]
r2_gd   = corr_gd ** 2

print(f"\n[방법 B: Full-field, grid interpolation]  grid={GRID_RES}x{GRID_RES}")
print(f"  RMSE={rmse_gd:.4f} V/m  MAE={mae_gd:.4f} V/m  MAPE={mape_gd:.2f}%  "
      f"Accuracy={acc_gd:.2f}%  R={corr_gd:.4f}  R^2={r2_gd:.4f}")

# ── 두 방법 교차검증 (critical check) ─────────────────────────────────────────
print("\n─── 두 방법 비교 (교차 검증) ───")
print(f"{'metric':10s} {'Point(A)':>12s} {'Grid(B)':>12s} {'diff':>10s}")
for name, a, b in [("RMSE", rmse_pt, rmse_gd), ("MAE", mae_pt, mae_gd),
                    ("MAPE%", mape_pt, mape_gd), ("Accuracy%", acc_pt, acc_gd),
                    ("R^2", r2_pt, r2_gd)]:
    print(f"{name:10s} {a:12.4f} {b:12.4f} {a - b:10.4f}")
if abs(acc_pt - acc_gd) > 5:
    print("[주의] 두 방법의 정확도 차이가 5%p 이상 — griddata 보간 스무딩 또는 "
          "mesh 해상도 차이로 인한 인공적 오차 가능성 점검 필요")
else:
    print("→ 두 방법이 서로 일관된 정확도를 보고함: 결과 신뢰 가능")


# ═══════════════════════════════════════════════════════════════════════════
# 종합 시각화 — 2행 3열
#  Row1: Full-field  [Python | COMSOL | Diff]
#  Row2: 진단        [Point diff% overlay | Parity plot | Error histogram]
# ═══════════════════════════════════════════════════════════════════════════
fig = plt.figure(figsize=(19, 12))
gs  = gridspec.GridSpec(2, 3, figure=fig, hspace=0.35, wspace=0.32)

VMIN, VMAX = 0, 0.05   # 조건 간 객관적 비교를 위해 colorbar range 고정
vmax_field = VMAX

ax1 = fig.add_subplot(gs[0, 0])
im1 = ax1.pcolormesh(Xi, Yi, python_grid, cmap="rainbow", vmin=VMIN, vmax=VMAX, shading="auto")
cb1 = fig.colorbar(im1, ax=ax1, shrink=0.85, label="AM$_{max}$ (V/m)")
apply_sci_colorbar(cb1, VMIN, VMAX)
ax1.set_title("Python Lead Field  AM$_{max}$", fontsize=17)

ax2 = fig.add_subplot(gs[0, 1])
im2 = ax2.pcolormesh(Xi, Yi, comsol_grid, cmap="rainbow", vmin=VMIN, vmax=VMAX, shading="auto")
cb2 = fig.colorbar(im2, ax=ax2, shrink=0.85, label="AM$_{vector}$ (V/m)")
apply_sci_colorbar(cb2, VMIN, VMAX)
ax2.set_title("COMSOL  AM$_{vector}$ (surface export)", fontsize=17)

# diff colorbar 범위를 field 범위(vmax_field)에 맞춤 → 두 필드가 유사한 스케일임이 시각적으로 드러남
ax3 = fig.add_subplot(gs[0, 2])
im3 = ax3.pcolormesh(Xi, Yi, diff_grid, cmap="RdBu_r", vmin=-vmax_field, vmax=vmax_field, shading="auto")
cb3 = fig.colorbar(im3, ax=ax3, shrink=0.85, label="Python - COMSOL (V/m)")
apply_sci_colorbar(cb3, -vmax_field, vmax_field)
ax3.set_title("Difference  (Python - COMSOL)", fontsize=17)
ax3.text(0.02, 0.02,
         f"RMSE={rmse_gd:.3f} V/m\nMAE={mae_gd:.3f} V/m\nMAPE={mape_gd:.2f}%\n"
         f"Accuracy={acc_gd:.2f}%\nR^2={r2_gd:.4f}",
         transform=ax3.transAxes, fontsize=14.5, va="bottom", ha="left",
         bbox=dict(boxstyle="round,pad=0.4", fc="white", ec="black", alpha=0.85))

for ax in (ax1, ax2, ax3):
    ax.set_xlabel("x (mm)", fontsize=15); ax.set_ylabel("y (mm)", fontsize=15)
    ax.set_aspect("equal")
    ax.grid(True, color="white", alpha=0.4, linewidth=0.6, linestyle="--", zorder=4)

# (2,1) Point-level diff% overlay
ax4 = fig.add_subplot(gs[1, 0])
triang_val = mtri.Triangulation(x, y)
tcf = ax4.tricontourf(triang_val, AM_max, levels=np.linspace(VMIN, VMAX, 51),
                      cmap="rainbow", vmin=VMIN, vmax=VMAX, extend="max")
cb4 = fig.colorbar(tcf, ax=ax4, shrink=0.85, label="AM$_{max}$ (V/m)")
apply_sci_colorbar(cb4, VMIN, VMAX)
diff_vmax_pt = np.nanpercentile(df_val["diff (%)"], 95)
sc = ax4.scatter(df_val["x (mm)"], df_val["y (mm)"], c=df_val["diff (%)"],
                  cmap="viridis", s=6, vmin=0, vmax=diff_vmax_pt, edgecolors="none", zorder=10)
fig.colorbar(sc, ax=ax4, shrink=0.85, pad=0.1, label="diff (%) [point-level]")
ax4.set_title("Point-level diff(%) overlay\n(nearest-node, AM_vector 좌표 전체)", fontsize=16)
ax4.set_xlabel("x (mm)", fontsize=15); ax4.set_ylabel("y (mm)", fontsize=15)
ax4.set_aspect("equal")
ax4.grid(True, color="white", alpha=0.4, linewidth=0.6, linestyle="--", zorder=4)

# (2,2) Parity plot — COMSOL vs Python (y=x)
ax5 = fig.add_subplot(gs[1, 1])
ax5.scatter(df_val["COMSOL_AM (V/m)"], df_val["Python_AM (V/m)"],
            s=4, alpha=0.25, color="steelblue", label=f"points (n={len(df_val)})")
lims = [0, max(df_val["COMSOL_AM (V/m)"].max(), df_val["Python_AM (V/m)"].max()) * 1.05]
ax5.plot(lims, lims, "k--", lw=1.2, label="y = x (perfect match)")
ax5.set_xlim(lims); ax5.set_ylim(lims)
ax5.set_xlabel("COMSOL AM_vector (V/m)", fontsize=15)
ax5.set_ylabel("Python AM_max (V/m)", fontsize=15)
ax5.set_title(f"Parity plot (point-level)\nR = {corr_pt:.4f}   R^2 = {r2_pt:.4f}", fontsize=16)
ax5.legend(fontsize=13, loc="upper left")
ax5.set_aspect("equal")
ax5.grid(True, alpha=0.3)

# (2,3) Error histogram
ax6 = fig.add_subplot(gs[1, 2])
ax6.hist(df_val["diff (%)"].dropna(), bins=60, color="indianred", alpha=0.8)
ax6.axvline(mape_pt, color="black", ls="--", lw=1.2, label=f"mean={mape_pt:.2f}%")
ax6.axvline(np.nanmedian(df_val["diff (%)"]), color="gray", ls=":", lw=1.2,
            label=f"median={np.nanmedian(df_val['diff (%)']):.2f}%")
ax6.set_xlabel("diff (%)", fontsize=15)
ax6.set_ylabel("count", fontsize=15)
ax6.set_title("Point-level 오차(%) 분포", fontsize=16)
ax6.legend(fontsize=13)
ax6.grid(True, alpha=0.3)

fig.suptitle(
    f"Lead Field vs COMSOL AM_vector -- 종합 검증  "
    f"(Point Accuracy={acc_pt:.2f}%, Full-field Accuracy={acc_gd:.2f}%)",
    fontsize=20, fontweight="bold", y=0.995,
)
plt.tight_layout()
val_save = SIM_DIR / f"Result_Efield_validation_{TAG}.png"
plt.savefig(val_save, dpi=150, bbox_inches="tight")
print(f"\nSaved: {val_save.resolve()}")
plt.show()
```

## 7. 전류 보존 검증 — 반지름별 flux(선속) 적분 비교

이론(공부 세션): 등방성 매질에서 점 전극이 만드는 전기장은 거리 r에 따라 크기가 변하지만,
그 전극을 감싸는 임의의 반지름 r에서 **각도로 적분한 flux는 r에 무관하게 일정**해야 한다.

$$\Phi(r) = \oint E_r(\theta)\cdot r\, d\theta \;\approx\; \text{const.} \;(\propto I/\sigma)$$

`E_f1[0]` (전극 1 단독, anti-phase 합성 이전) 필드로 실제 COMSOL FEM 데이터에서 이 관계를 검증한다.


```python
# ── 전극 위치: 4번 셀(ELECTRODE_XY_TABLE/ELECTRODES)에서 정의한 실측 좌표 사용 ──
E_single = E_f1[0]                       # (N,2) 전극1(CH1 top) 단독 필드 [V/m] @ 1mA (anti-phase 합성 이전)
E_mag_single = np.linalg.norm(E_single, axis=1)

x0, y0 = ELECTRODES["CH1 top"]           # f1 파일의 전극1 = CH1 top, 실측 좌표
print(f"전극1(CH1 top) 위치: ({x0:.2f}, {y0:.2f}) mm  (ELECTRODE_XY_TABLE 기반, argmax 추정 아님)")

dx, dy = coords[:, 0] - x0, coords[:, 1] - y0
r_all     = np.hypot(dx, dy)
theta_all = np.arctan2(dy, dx)            # 전극 중심 기준 각도 [rad]

_trapz = getattr(np, "trapezoid", None) or np.trapz   # numpy>=2.0 호환 (trapz -> trapezoid)
```

### 7-1. Flux 계산 함수

반지름 `r_target` 근방(두께 `dr`)의 mesh 노드를 골라, 전기장의 반지름 방향 성분
$E_r = \mathbf{E}\cdot\hat{n}$ ($\hat{n}=(\cos\theta,\sin\theta)$) 을 계산하고,
`theta`로 정렬한 뒤 $\displaystyle\Phi(r)=\int E_r(\theta)\, r\, d\theta$ 를 적분(`_trapz`)한다.

전극이 도메인 경계 위에 있어 데이터가 존재하는 각도 범위(반원 근처)만 자동으로 반영된다(0으로 채우지 않음 — 있는 노드만 사용).


```python
def flux_at_radius(r_target, dr, coords, E_field, x0, y0, r_all, theta_all):
    """
    반지름 r_target 근방(두께 dr)에서 flux Phi(r) = 반지름 방향 E성분을 각도로 적분한 값을 계산.
    r_target, dr: [mm]
    E_field: (N,2) 전극 단독 필드
    반환: flux 값 (float), 사용된 노드 수 (int)
    """
    mask = (r_all >= r_target - dr / 2) & (r_all <= r_target + dr / 2)
    n = mask.sum()
    if n < 3:
        return 0.0, n

    theta_sel = theta_all[mask]
    order = np.argsort(theta_sel)
    theta_sorted = theta_sel[order]

    n_hat = np.stack([np.cos(theta_sorted), np.sin(theta_sorted)], axis=1)
    E_sel = E_field[mask][order]
    E_r_sorted = np.sum(E_sel * n_hat, axis=1)   # 반지름 방향 성분

    flux = _trapz(E_r_sorted * r_target, theta_sorted)
    return flux, n


# 검증: 여러 반지름에서 flux가 거의 일정한지 비교 (도메인 크기에 맞게 값 조정 가능)
for r_test in [3, 5, 8, 12, 18]:
    flux, n_pts = flux_at_radius(r_test, dr=1.0, coords=coords, E_field=E_single,
                                  x0=x0, y0=y0, r_all=r_all, theta_all=theta_all)
    print(f"r={r_test:5.1f}mm  flux={flux:10.4f}  (사용 노드 {n_pts}개)")
```
